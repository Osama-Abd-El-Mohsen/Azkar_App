```python
import pandas as pd

data = pd.read_csv('azkar.csv',index_col=0)
data.head()
```

```python
interested_cats =[
    "أذكار الصباح",
    "أذكار المساء",
    "سورة الملك",
    "الذكر قبل الوضوء",
    "الذكر بعد الفراغ من الوضوء",
    "دعاء الاستفتاح",
    "الأذكار بعد السلام من الصلاة",
    "أذكار الاستيقاظ من النوم",
    "الاستغفار و التوبة",
    "أذكار النوم",
    "الذكر عند دخول المنزل",
    "الذكر عند الخروج من المنزل",
    "دعاء الهم والحزن",
    "دعاء الكرب",
    "دعاء من استصعب عليه أمر",
    "دعاء الوسوسة في الصلاة و القراءة",
    "دعاء طرد الشيطان و وساوسه",
    "دعاء من أصيب بمصيبة",
    "دعاء صلاة الاستخارة",
    "الدعاء عند إفطار الصائم",
    "الدعاء يوم عرفة",
    "الرُّقية الشرعية من القرآن الكريم",
    "الرُّقية الشرعية من السنة النبوية",
    ]

filtered_data = data[data["category"].isin(interested_cats)]
len(filtered_data.category.unique())
filtered_data.category.unique()
filtered_data
```

```python
filtered_data['category'] = pd.Categorical(filtered_data['category'], categories=interested_cats, ordered=True)
df_sorted = filtered_data.sort_values(['category', 'index'])
df_sorted.to_csv('azkar.csv', encoding="utf-8-sig")
```

```python
cat_data = data[data.category == 'أذكار الصباح']

zekr_list = []
description_list=[]
count_list = []
reference_list=[]

def get_cat_data(row):
    for index,field in enumerate(row.index):
        if index == 1 :
            zekr_list.append(row[field])
        elif index == 2 :
            description_list.append(row[field])
        elif index == 3 :
            count_list.append(row[field])
        elif index == 4 :
            reference_list.append(row[field])

cat_data.apply(get_cat_data, axis=1)
print(zekr_list)
print(description_list)
print(count_list)
print(reference_list)
```

```python
azkar_data = pd.read_csv('azkar.csv',index_col=0)

zekr_list = []
description_list=[]
count_list = []
reference_list=[]
def get_cat_data(row):

    for index,field in enumerate(row.index):
        if index == 1 :
            zekr_list.append(row[field])
        elif index == 2 :
            description_list.append(row[field])
        elif index == 3 :
            count_list.append(row[field])
        elif index == 4 :
            reference_list.append(row[field])

def get_azkar(zekr_cat:str):
    cat_data = azkar_data[azkar_data.category == zekr_cat]
    cat_data.apply(get_cat_data, axis=1)

    print(zekr_list)
    print(description_list)
    print(count_list)
    print(reference_list)
get_azkar("أذكار الصباح")
```
